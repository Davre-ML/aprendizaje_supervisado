```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic_clean.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por clase y supervivencia
conteo = df.groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)

# Preparar datos
clases = conteo.index
no_sobrevivio = conteo[0]
sobrevivio = conteo[1]

# Crear figura
plt.figure(figsize=(8, 6))

# Ancho de barra y posiciones
bar_width = 0.35
x = range(len(clases))

# Barras
plt.bar([pos - bar_width/2 for pos in x], no_sobrevivio, width=bar_width,
        label="No Sobrevivi", color="#333333")  # gris oscuro
plt.bar([pos + bar_width/2 for pos in x], sobrevivio, width=bar_width,
        label="Sobrevivi", color="#5555AA")  # azul oscuro

# Etiquetas y ttulo
plt.xlabel("Clase del Pasajero (Pclass)")
plt.ylabel("Recuento de Pasajeros")
plt.title("Tasa de Supervivencia por Clase")
plt.xticks(x, clases)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Agrupa las edades en categoras (Infante, Nio, Adolescente, Joven Adulto, Adulto, Mayor) y crea una grfica que muestre la tasa de supervivencia por grupo de edad.
import pandas as pd
import matplotlib.pyplot as plt

# Definir bins y etiquetas
bins = [0, 2, 12, 17, 29, 59, 120]
labels = ["Infante", "Nio", "Adolescente", "Joven Adulto", "Adulto", "Mayor"]

# Crear columna con categoras de edad
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Calcular la tasa de supervivencia por grupo de edad
tasa_supervivencia = df.groupby('AgeGroup')['Survived'].mean()

# Crear grfica
plt.figure(figsize=(10, 6))
plt.bar(tasa_supervivencia.index, tasa_supervivencia.values)

# Etiquetas y ttulo
plt.xlabel("Grupo de Edad")
plt.ylabel("Tasa de Supervivencia")
plt.title("Tasa de Supervivencia por Grupo de Edad")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Encuentra y muestra los valores mximo y mnimo de la columna de edad.
# Encontrar valores mnimo y mximo de edad
edad_minima = df['Age'].min()
edad_maxima = df['Age'].max()

print("Edad mnima:", edad_minima)
print("Edad mxima:", edad_maxima)
```

```python
#Crea una grfica que muestre la distribucin de edades, incluyendo lneas verticales que indiquen la media, mediana y moda de las edades.
import matplotlib.pyplot as plt
import pandas as pd

# Eliminar valores nulos de edad
edades = df['Age'].dropna()

# Calcular estadsticas
media = edades.mean()
mediana = edades.median()
moda = edades.mode()[0]   # mode() puede devolver varias, tomamos la primera

# Crear figura
plt.figure(figsize=(10, 6))

# Histograma
plt.hist(edades, bins=30)

# Lneas verticales
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f"Media: {media:.2f}")
plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f"Mediana: {mediana:.2f}")
plt.axvline(moda, color='blue', linestyle='-.', linewidth=2, label=f"Moda: {moda:.2f}")

# Etiquetas y ttulo
plt.xlabel("Edad")
plt.ylabel("Cantidad de Pasajeros")
plt.title("Distribucin de Edades con Media, Mediana y Moda")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Crea un diagrama de caja (boxplot) que muestre los outliers en la columna de edad.
import matplotlib.pyplot as plt

# Eliminar valores nulos de edad
edades = df['Age'].dropna()

# Crear figura
plt.figure(figsize=(8, 6))

# Boxplot
plt.boxplot(edades, vert=True)

# Etiquetas y ttulo
plt.ylabel("Edad")
plt.title("Diagrama de Caja de la Edad (con Outliers)")

plt.tight_layout()
plt.show()
```

```python
#Crea una grfica que muestre la distribucin de tarifas, incluyendo lneas verticales que indiquen la media, mediana y moda de las tarifas.
import matplotlib.pyplot as plt
import pandas as pd

# Eliminar valores nulos de tarifa
tarifas = df['Fare'].dropna()

# Calcular estadsticas
media = tarifas.mean()
mediana = tarifas.median()
moda = tarifas.mode()[0]   # mode() puede devolver varias; tomamos la primera

# Crear figura
plt.figure(figsize=(10, 6))

# Histograma
plt.hist(tarifas, bins=30)

# Lneas verticales
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f"Media: {media:.2f}")
plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f"Mediana: {mediana:.2f}")
plt.axvline(moda, color='blue', linestyle='-.', linewidth=2, label=f"Moda: {moda:.2f}")

# Etiquetas y ttulo
plt.xlabel("Tarifa (Fare)")
plt.ylabel("Cantidad de Pasajeros")
plt.title("Distribucin de Tarifas con Media, Mediana y Moda")

plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Crea un mapa de calor (heatmap) que muestre la matriz de correlacin de las caractersticas numricas en el dataset.
import matplotlib.pyplot as plt

# Filtrar solo columnas numricas
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlacin
corr_matrix = df_numeric.corr()

# Crear el heatmap
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, interpolation='nearest')
plt.colorbar()

# Etiquetas
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

plt.title("Matriz de Correlacin (Titanic)")
plt.tight_layout()
plt.show()
```
